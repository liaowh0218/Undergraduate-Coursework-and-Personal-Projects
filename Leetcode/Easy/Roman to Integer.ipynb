```python
class Solution:
    def romanToInt(self, s: str) -> int:
        dict = {'I':1, 'V':5, 'X':10, 'L':50, 'C':100, 'D':500, 'M':1000}
        total = 0
        previous = 0
        for i in reversed(s):
            if dict[i] >= previous:
                total += dict[i]
                previous = dict[i]
            else:
                total -= dict[i]
                previous = dict[i]
        return total
```

```python
Solution().romanToInt('MCMXCIV') #1994
```

```python
Solution().romanToInt('III') #3
```
