```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        if x < 0:
            return False
        a = str(x)
        if a[:len(a)//2] == a[::-1][:len(a)//2]:
            return True
        else:
            return False
```

```python
Solution().isPalindrome(11)
```

```python
Solution().isPalindrome(121)
```
